## Dataframeの列と行の選択: iloc・locメソッド

- ilocメソッド?
  - 行・列を数字INDEXで選択
  - **iloc[row、col]**
  - Slicing 可能。
- locメソッド
  - row, colをINDEXの名前で選択
  - **loc[2:4 , 'row_name' : 'col_name']**
  -  Slicing 可能

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./datasets/bike_rentals/bike_rentals.csv')
```

```python
df.iloc[2, 3]
df.head(3)
```

```python
# df.iloc[2:5, 3:6]
df.loc[2:4, ['workingday' , 'temp', 'casual']]
```

###　locを利用した条件探索
- **locメソッド**
  - loc[条件]を選択するには、loc[条件, row_name]で使う。

```python
# dfでseasonカラムで値が2のやつを出力したい。
df.loc[df['season'] == 2]
# dfでseasonカラムで値が2で、casualから出力
df.loc[df['season'] == 2, 'casual':]
#条件が多くなると query()を使ったほうがいい。
df.query("season == 2")[['casual', 'registered', 'count']]
```

```python
df.loc[(df['season'] != 1) & (df['weather'] != 2)]
```
